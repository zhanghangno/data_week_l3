```python
#Thinking1：什么是监督学习，无监督学习，半监督学习
"""
 1.监督学习是利用已知数据和对应标签，训练一个算法，得到输入数据与输出数据的对应关系。常见监督学习算法有：线性回归，逻辑回归，决策树，朴素贝叶斯，knn算法
 2.无监督学习是已知数据没有标签，也没有确定的结果，常见无监督学习的算法有k-means聚类算法，
 3.半监督学习介于监督学习与无监督学习之间，其任务与监督学习一致，任务中包含明确的目标，采用数据有含有标签的数据，和无标签的数据。
"""
```

```python
#Thinking3：随机森林采用了bagging集成学习，bagging指的是什么
"""
bagging是指将若干个弱分类器得分类结果进行投票选择，从而组成一个强分类器，且bagging不用单课棵决策树来做预测，增加了预测得正确率，及不容易出现过拟合，减少而预测方差
"""
```

```python
#Thinking4：主动学习和半监督学习的区别是什么
"""
主动学习，再利用我未来标注得时候，是从未标记得数据中找到最容易判断得样例交由人工进行标注，这个过程是筛选差数据得过程，也是一个交互得过程。
半监督学习，对于为标记数据而言，是选择最不容易判断错误得样例加入 到已标注数据中，这个过程，是自动得过程，是筛选数据做好得过程，
不需要人工干预，基于自身对未标记数据加以利用，提高学习模型得泛化性能。
"""
```

```python
#加载停用词
with open(r"C:\Users\adminPC\Desktop\数据挖掘算发资料\第三天资料\L3\L3\Assignment\chinese_stopwords.txt","r",encoding="utf-8") as file:
    stopwords = [i[:-1] for i in file.readlines()]

def split_text(text):
    text = text.replace(' ','').replace("\n","")
    text2 = jieba.cut(text)
    result = ' '.join([w for w in text2 if w not in stopwords])
    return result
    
```

```python
#计算corpus de TFIDF
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
conterverctorizer = CountVectorizer(encoding="gb18030",min_df=0.015)
tfidftransformer = TfidfTransformer()
countvector = conterverctorizer.fit_transform(corpus)
tfidf = tfidftransformer.fit_transform(countvector)
print(tfidf.shape)
```

```python
#是否标记为自己得新闻
label =list(map(lambda source: 1 if "新华社" in str(source) else 0,news.source))
label
             
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,recall_score,precision_score
X_train,X_test,y_train,y_test =train_test_split(tfidf.toarray(),label,test_size=0.3)
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
model = MultinomialNB()
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
print("准确率:",accuracy_score(y_test,y_predict))
print("精确率:",precision_score(y_test,y_predict))
print("召回率:",accuracy_score(y_test,y_predict))
```

```python
#使用模型进行风格预测
prediction = model.predict(tfidf.toarray())
labels = np.array(label)
#compare_news_index 有两种风格，prediction 为预测风格， label为预测风格
compare_news_index = pd.DataFrame({"prediction":prediction,"labels":labels})
copy_nwes_index = compare_news_index[(compare_news_index["prediction"]==1) & (compare_news_index["labels"]==0).index ]
```

```python
 #实际为新华社的新闻
xinhuashe_news_index = compare_news_index[(compare_news_index["labels"]==1)].index
```

```python
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans

normalizer = Normalizer()
scaled_aarray = normalizer.fit_transform(tfidf.toarray())

kmeans = KMeans(n_clusters=25)
k_label = kmeans.fit_predict(scaled_aarray)
print(k_label.shape)
```

```python
from collections import defaultdict
#创建索引
id_class ={index:class_ for index,class_ in enumerate(k_label)}
class_id = defaultdict(set)
for index,class_ in id_class.items():
    if index in xinhuashe_news_index.tolist():
        class_id[class_].add(index)
        
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#查找相似的文章
def find_similar_text(cpindex,top=10): 
    dist_dict = {i:cosine_similarity(tfidf[cpindex],tfidf[i]) for i in class_id[id_class[cpindex]]}
    #从大到小排序
    return sorted(dist_dict.items(),key= lambda x:x[1][0],reverse=True)[:top]
```

```python
similar_list  = find_similar_text(cpindex)
print(similar_list)
```
